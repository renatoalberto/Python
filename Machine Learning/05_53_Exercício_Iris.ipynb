```python
from sklearn.datasets import load_iris
iris = load_iris()
x = iris.data
y = iris.target
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
from sklearn import metrics
```

```python
resultados_previsoes = {}

for i in range(25):
    knn = KNeighborsClassifier(n_neighbors=i+1)
    
    knn.fit(x_train, y_train,)
    
    previsoes = knn.predict(x_test)    
    
    acertos = metrics.accuracy_score(y_test, previsoes)

    # print(f'Compara resultados para {i+1} visinho(s) obteve {round(acertos,4)}% de acertos:\n{y_test}\n{previsoes}')    
    
    resultados_previsoes[i+1] = round(acertos,4)
```

```python
import matplotlib.pyplot as plt

plt.plot(list(resultados_previsoes.keys()), list(resultados_previsoes.values()))
plt.xlabel('Valores de K')
plt.ylabel('Performance')
```
