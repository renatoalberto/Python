### Histograma e Gráfico de Pizza

```python
# Recuperando o dataset 'titanic'
import pydataset

titanic = pydataset.data('titanic')
titanic.head()   # head() trans os 5 primeiros
```

```python
# O titanic é basicamente um conjunto de dados categóricos

# Identificando as classes possíveis e quantidades de cada
print('class:\n', titanic['class'].value_counts(), sep='', end='\n\n')
# Identificando idade possíveis e quantidades de cada
print('age:\n', titanic['age'].value_counts(), sep='', end='\n\n')
# Identificando sexos possíveis e quantidades de cada
print('sex:\n', titanic['sex'].value_counts(), sep='', end='\n\n')
# Identificando se sobrevivei e quantidades de cada
print('survived:\n', titanic['survived'].value_counts(), sep='', end='\n\n')
```

```python
# Pandas possui um atalho para plotar um gráfico
titanic['class'].value_counts().plot(kind='bar')
plt.show()
```

```python
# Pandas possui um atalho para plotar um gráfico
titanic['survived'].value_counts().plot(kind='bar')
plt.show()
```

```python
# Para apresentar quantidade por sobreviveu e classe
titanic.groupby('survived')['class'].value_counts()
```

```python
# Com dados agrupados é só plotar
titanic.groupby('survived')['class'].value_counts().plot(kind='bar')
plt.show()
```

```python
# carregando arquivo csv da população brasileira
import pandas as pd

estados = pd.read_csv('populacao_brasil.csv')
estados.head()
```

```python
# Gerando um histograma de barras vesticais e totais em anotação científica
estados['total'].hist()
plt.show()
```

```python
# com a função subplots, que retorna a figura e eixos para ter um controle maior sobre o gráfico
fig, ax = plt.subplots()

# bins para tamanho do intervalos
# orientation para orientação do gráfico
plt.hist(estados['total'], bins=15, orientation='horizontal') 

# ticklabel_format para apresentar cada total no eixo, como está no arquivo
ax.ticklabel_format(style='plain')

plt.show()
```

```python
# Criando a coluna percentual
estados['percentual'] = estados['total'] / estados['total'].sum()
estados.head()
```

```python
estados['percentual'].sum()
```

```python
# Gráfico de pizza é o pie()
# autopct para incluir os percentuais na pizza
plt.pie(estados['percentual'], labels=estados['estado'], autopct='%1.2f%%')
plt.show()
```

```python
# Pegando dataset Iris da biblioteca pydataset
df_iris = pydataset.data('iris')
df_iris.head()
```

```python
print(df_iris['Species'].unique())
```

```python
# Aplicando um gráfico de dispersão com 3 variáveis
# Eixo x - df_iris['Sepal.Length']
# Eixo y - df_iris['Sepal.Width']
# Tamanho do ponto - df_iris['Petal.Length']
plt.scatter(df_iris['Sepal.Length'], df_iris['Sepal.Width'], sizes=20 * df_iris['Petal.Length'])
plt.show()
```

```python
# Aplicando cor ao ponto do gráfico de dispersão anterior
def specie_color(x):
    if   x == 'setosa':
        return 0
    elif x == 'versicolor':
        return 1
    elif x == 'virginica':
        return 2

df_iris['SpeciesNumber'] = df_iris['Species'].apply(specie_color)

df_iris.head()
```

```python
# Aplicando a cor de acordo com a espécie
# cmap  - paleta de cores: 'viridis', 'plasma', 'inferno', 'magma'
# alpha - percentual de transparência
plt.scatter(
    df_iris['Sepal.Length'], 
    df_iris['Sepal.Width'], 
    sizes=20 * df_iris['Petal.Length'],
    c=df_iris['SpeciesNumber'], cmap='viridis', alpha=0.4
)
plt.show()
```

### Customizando Eixos e Labels de um Gráfico

```python
# Dataset reported contendo crimes suécos de 1950 - 2015
# https://www.kaggle.com/mguzmann/swedishcrime
df_reported = pd.read_csv('reported.csv')
df_reported.head()
```

```python
# Gráfico simples não permite identificar x e y
plt.plot(df_reported['Year'], df_reported['crimes.total'], '-r')
plt.plot(df_reported['Year'], df_reported['crimes.person'], '-b')
plt.show()
```

```python
# Gráfico com legendas
# utilizando subplot para maior controle
fig, ax = plt.subplots()

ax.plot(df_reported['Year'], df_reported['crimes.total'], '-r', label='crimes.total')
ax.plot(df_reported['Year'], df_reported['crimes.person'], '-b', label='crimes.person')

ax.legend(loc='upper left')

ax.set_ylabel('Quantidade')
ax.set_xlabel('Ano')

ax.set_title('Crimes: Total x Pessoa')

# Ajustando o limite mínimo e máximo de x
ax.set_xlim([df_reported['Year'].min(), df_reported['Year'].max()])

plt.show()
```

### Gráficos de Séries Temporais

```python
# Dados sobre visualizações de uma página por hora
df_visualizacoes = pd.read_csv('ppz-jan-fev-2017.csv')
df_visualizacoes.head()
```

```python
# Horas contadas a partir do dia 17/01/01
import datetime as dt

def to_date(x):
    return dt.datetime(2017, 1, 1) + dt.timedelta(hours=x)

df_visualizacoes['data'] = df_visualizacoes['hour'].apply(to_date)

df_visualizacoes.head()
```

```python
# Colocando coluna data como indice e deletando coluna hour
df_visualizacoes.set_index(['data'], inplace=True)
del df_visualizacoes['hour']
df_visualizacoes.head()
```

```python
df_visualizacoes.index.day.min()
```

```python
# Criando o gráfico plt_date()
import matplotlib.dates as dates

fig, ax = plt.subplots()
ax.plot_date(df_visualizacoes.index.to_pydatetime(), df_visualizacoes['views'], 'b-')
ax.xaxis.set_minor_locator(dates.DayLocator(bymonthday=pd.RangeIndex(5, 32, 5)))
ax.xaxis.set_minor_formatter(dates.DateFormatter('%d'))
ax.xaxis.grid(True, which='minor')

ax.yaxis.grid()
ax.xaxis.set_major_locator(dates.MonthLocator())
ax.xaxis.set_major_formatter(dates.DateFormatter('%b'))  # %b com 3 caracteres

plt.tight_layout()
plt.show()
```

```python
# Tratando locale
import locale

locale.setlocale(locale.LC_ALL, 'pt-BR')

fig, ax = plt.subplots()

ax.plot_date(df_visualizacoes.index.to_pydatetime(), df_visualizacoes['views'], 'b-')

ax.xaxis.set_minor_locator(dates.DayLocator(bymonthday=pd.RangeIndex(5, 32, 5)))
ax.xaxis.set_minor_formatter(dates.DateFormatter('%d'))

ax.xaxis.set_major_locator(dates.MonthLocator())
ax.xaxis.set_major_formatter(dates.DateFormatter('%b'))  # %b mês com 3 caracteres

ax.xaxis.grid(True, which='minor')
ax.yaxis.grid()

plt.tight_layout()
plt.show()
```

```python
# total_bill - total da conta
# tip - gorjeta
# size - quantidade de pessoas
tips.head()
```

### distplot - faz um histograma e plota uma curva KDE

```python
# Vizualização da distribuição do total da conta
# Curva KDE - Estimativa de densidade kernel, para retirar informar kde=False
# bins - quantidade de colunas da distribuição
sns.distplot(tips['total_bill'], bins=50)
```

### kdeplot - apresenta a curva KDE

```python
sns.kdeplot(tips['total_bill'])
```

### joinplot - faz um histograma de duas variáveis ao mesmo tempo

```python
# kind para tipo de gráfico:
#   *scatter - padrão
#   *reg     - regressão linear
#   *hex     - mapa de calor onde conjunto de dados maior
sns.jointplot(x='total_bill', y='tip', data=tips, kind='reg')
```

### pairplot - faz joinplot para cada coluna numérica

```python
# use parâmetro hue para distiguir com cores algum dado categórico
sns.pairplot(tips, hue='sex')
```

### rugplot - assemelha ao distplot demonstrando onde há maior a concentraçao 

```python
sns.rugplot(tips['total_bill'])
sns.kdeplot(tips['total_bill']) # Confirmando a maior concentração
```

```python
sns.stripplot(x='day', y='total_bill', data=tips, jitter=True)
```

```python
# hue para quebra distribuição entre homens e mulheres
sns.stripplot(x='day', y='total_bill', data=tips, hue='sex', jitter=True)
```

### swarmplot se assemelha a stripplot mas trabalha melhor a sobreposição de pontos

```python
# Atenção que ele lida mal com grandes quantidade de dados
sns.swarmplot(x='day', y='total_bill', data=tips)
```

```python
# hue para distinguir pontos homem e mulher
sns.swarmplot(x='day', y='total_bill', data=tips, hue='sex')
```

```python
# dodge, mesma função do split, e separa os pontos
sns.swarmplot(x='day', y='total_bill', data=tips, hue='sex', dodge=True)
```

#### Juntando swarmplot e violinplot para visualizar onde estão os pontos nas curvas KDE

catplot - apresenta qualquer dos gráficos anteriores de acordo com parâmetro kind

```python
# flights - dataset sobre número de passageiros num companhia aérea
flights = sns.load_dataset('flights')
flights.head()
```

```python
# corr() - Matriz de correlação dos dados categóricos 
tips.corr()
```

### heatmap - mapa de calor

```python
# cmap - paleta de cores
# annot - descrição de valor de cada quadrante
# linecolor - cor da linha que separa os quadrante
# linewidths - largura da linha que separa os quadrantes
sns.heatmap(tips.corr(), cmap='coolwarm', annot=True, linecolor='white', linewidths=1)
```

```python
# Criando um pivot_table que melhora a viasualização e disposição dos dados
pivtab_flights = flights.pivot_table(values='passengers', index='year', columns='month')
pivtab_flights
```

```python
# Fica claro com o mapa de calor que com o passar dos anos a quantidade de passageiros tem aumentado
sns.heatmap(pivtab_flights, cmap='coolwarm')
```

### clustermap - auxilia visualização de padrões nos dados

```python
# standard_scale - limita tamanho da escala e aproxima as cores
sns.clustermap(pivtab_flights, standard_scale=1)
```

### lmplot - linear model plot, com um gráfico de dispersão e um modelo linear

```python
# Visualizando a relação entre total da conta e a gorjeta
# hue - para segregar sobre algum dado categórico
# palette - para paleta de cores
# markers - tipo de marcadores (o-bolinha, v-triangulo)
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', palette='hot', markers=['o', 'v'])
```

```python
# col - separar a plotagem 
sns.lmplot(x='total_bill', y='tip', data=tips, col='sex', hue='sex')
```

```python
# row - para segregação em linha de acordo com o dado categórico informado
sns.lmplot(x='total_bill', y='tip', data=tips, col='sex', hue='sex', row='time')
```

```python
# aspect - proporção de cada gráfico
# height - altura do gráfico
sns.lmplot(x='total_bill', y='tip', data=tips, col='day', hue='sex', aspect=0.7, height=6)
```
