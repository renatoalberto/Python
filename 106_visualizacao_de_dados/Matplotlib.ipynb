```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
salarios = np.array([100, 200, 300, 500, 250, 150])
```

```python
# Forma básica
plt.plot(salarios)
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Título')
plt.show()
```

```python
# color ou c      - cores, aceita hexadecimal
# linewidth ou lw - lagura da linha
# alpha           - transparência da cor
# linestyle ou ls - estilo da linnha
plt.plot(salarios, c='#FF00FF', ls='-.', linewidth='10', alpha=0.5)
plt.show()
```

```python
# Renderizando marcadores
plt.plot(salarios, marker='s')
plt.show()
```

```python
# Renderizando marcadores com tamanho 18
plt.plot(salarios, marker='s', ms=18)
plt.show()
```

```python
# Comparando no Gráfico
salarios_marcos = np.array([100, 200, 300, 500, 250, 150])
salarios_gileno = np.array([50, 300, 450, 300, 650, 400])
```

```python
plt.plot(salarios_marcos, marker='o', c='Black', ls='--', label="Marcos")
plt.plot(salarios_gileno, marker='x', c='Red', label="Gileno")
plt.legend()
plt.show()
```

```python
#Apresentando a legenda na posição inferior direita
"""
Posições Possíveis:
	best
	upper right
	upper left
	lower left
	lower right
	right
	center left
	center right
	lower center
	upper center
	center
"""
plt.plot(salarios_marcos, marker='o', c='Black', ls='--', label="Marcos")
plt.plot(salarios_gileno, marker='x', c='Red', label="Gileno")
plt.legend(loc='lower right')
plt.show()
```

```python
data_iris = np.genfromtxt('iris.data', delimiter=",", usecols=(0, 1, 2, 3))
```

```python
# Apresentando um título
plt.plot(data_iris[:50, 0]    , marker="o", ms="3", c="Red"  , ls=":", label="Iris-setosa")
plt.plot(data_iris[50:100, 0] , marker="o", ms="3", c="Green", ls=":", label="Iris-versicolor")
plt.plot(data_iris[100:150, 0], marker="o", ms="3", c="Blue" , ls=":", label="Iris-virginica")
plt.title('Comprimento da Sépala')
plt.legend()
plt.show()
```

### Visualizando a curva da função seno()

```python
# Gerando 100 valores com interavos iguais com a função linspace() do numpy
x = np.linspace(1, 10, 100)
x
```

```python
# r-- para linha red e pontilhada
plt.plot(x, np.sin(x), 'r--')
plt.show()
```

```python
# Para salvar uma imagem com savefig()
fig = plt.figure()
plt.plot(x, np.sin(x), 'r--')
plt.show()
fig.savefig('sin-01.png')
```

```python
# subplots para plotagem de mais de um gráfico por linha
plt.subplots(nrows=1, ncols=2)
```

```python
# subplot para definir qual o quadrant
plt.subplot(1, 2, 1)
plt.plot(salarios_marcos, 'r--')

plt.subplot(1, 2, 2)
plt.plot(salarios_gileno, 'g*-')

plt.show()
```

```python
# Modo comum utilizando figure, com mais controle sobre posição dos gráficos
fig = plt.figure()      # Figure é uma caixa vazia onde plotamos os gráficos

# Criando um eixo
# 1ª argumento percentual distancia esquerda
# 2ª argumento percentual distancia baixo
# 3ª argumento percentual largura
# 3ª argumento percentual altura
axes1 = fig.add_axes([0.1, 0.1, 1, 0.8]) 

axes1.plot(salarios_marcos)
axes1.set_xlabel('Mês')
axes1.set_ylabel('Salário')
axes1.set_title('Salário Marcos')

axes2 = fig.add_axes([1.3, 0.1, 1, 0.8]) # Repare na distância esquerda

axes2.plot(salarios_gileno)
axes2.set_xlabel('Mês')
axes2.set_ylabel('Salário')
axes2.set_title('Salário Gileno')

plt.show()
```

```python
# Subsplots com "s" no final
fig, ax = plt.subplots()

ax.plot(salarios_gileno, 'b--')
ax.plot(salarios_marcos, 'g-.')

ax.set_xlabel('Mês')
ax.set_ylabel('Salário')
ax.set_title('Comparação entre Salários')

plt.show()
```

```python
# Subsplots com "s" no final passando parâmetros de linhas e colunas
fig, ax = plt.subplots(nrows=1, ncols=2)  # nesse caso ax recebe uma lista de eixos

ax[0].plot(salarios_gileno, 'b--')
ax[0].set_xlabel('Mês')
ax[0].set_ylabel('Salário')
ax[0].set_title('Salários Gileno')

ax[1].plot(salarios_marcos, 'g-.')
ax[1].set_xlabel('Mês')
ax[1].set_ylabel('Salário')
ax[1].set_title('Salários Marcos')

plt.tight_layout()  # Método que cuidará das sobreposições

plt.show()
```

```python
# figsize para ajuste do tamanho da imagem
# dpi para resolução da imagem
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(10, 4), dpi=200)  

ax[0].plot(salarios_gileno, 'b--')
ax[0].set_xlabel('Mês')
ax[0].set_ylabel('Salário')
ax[0].set_title('Salários Gileno')

ax[1].plot(salarios_marcos, 'g-.')
ax[1].set_xlabel('Mês')
ax[1].set_ylabel('Salário')
ax[1].set_title('Salários Marcos')

plt.tight_layout()  # Método que cuidará das sobreposições

plt.show()
```

```python
# Incluindo label
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(10, 4), dpi=200)  

ax[0].plot(salarios_gileno, 'b--', label="Gileno") # Incluir texto label
ax[0].set_xlabel('Mês')
ax[0].set_ylabel('Salário')
ax[0].set_title('Salários Gileno')
ax[0].legend()                                     # Ativa a legenda

ax[1].plot(salarios_marcos, 'g-.', label="Marcos")
ax[1].set_xlabel('Mês')
ax[1].set_ylabel('Salário')
ax[1].set_title('Salários Marcos')
ax[1].legend(loc=2)    # loc para definir o local do label

plt.tight_layout()  

plt.show()
```

```python
# Ajustando limites do gráfico com set_xlim e set_ylim
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(10, 4), dpi=200)  

ax[0].plot(salarios_gileno, 'b--', label="Gileno") # Incluir texto label
ax[0].set_xlabel('Mês')
ax[0].set_ylabel('Salário')
ax[0].set_title('Salários Gileno')
ax[0].set_ylim([0, 800])
ax[0].set_xlim([0,  5])
ax[0].legend()                                    

ax[1].plot(salarios_marcos, 'g-.', label="Marcos")
ax[1].set_xlabel('Mês')
ax[1].set_ylabel('Salário')
ax[1].set_title('Salários Marcos')
ax[1].set_ylim([0, 800])
ax[1].set_xlim([0,  5])
ax[1].legend(loc=2)    

plt.tight_layout()  

plt.show()
```
