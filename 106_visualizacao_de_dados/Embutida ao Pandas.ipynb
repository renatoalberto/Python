### plot.line() - gráfico de linha

```python
df1.plot.line(y='B', figsize=(12,5), lw=1)
```

### plot.scatter() - gráfico de dispersão

```python
# Gráfico de dispersão A em relação ao B
# Onde o parâmetro C fará a distinção por cores
# cmap para paleta de cores
df1.plot.scatter(x='A', y='B', c='C', cmap='inferno')
```

### plot.box() - Gráfico de caixa para visualização de grupos por meio de quartis

```python
df2.plot.box()
```

```python
df = pd.DataFrame(np.random.rand(1000, 2), columns=['A', 'B'])
df.head()
```

### plt.hexbin() -  gráfico de categorização hexagonal de x versus y.

```python
# Gráfico hexagonal
# gridsize para tamanho dos hexagonos
df.plot.hexbin(x='A', y='B', gridsize=20, cmap='Oranges')
```
