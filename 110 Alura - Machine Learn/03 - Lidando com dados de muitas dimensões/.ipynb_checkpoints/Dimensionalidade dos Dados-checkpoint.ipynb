### Problemas de bases com alta dimensionalidade?
Para os algoritmos de Machine Learning, quanto maior o número de dimensões, mais lento é o processo e mais espaço de armazenamento é necessário. Além disso, a alta dimensionalidade pode superajustar os nossos modelos aos dados do treinamento.

A Alura Care possui uma base de dados contendo o diagnóstico do tipo de câncer, que pode ser maligno (M) ou benigno (B), e diversos exames/dimensões que caracterizam esse diagnóstico.</br></br>

### Avaliando com um modelo DummyClassifier
Se não houvesse modelo nenhum, qual seria meu grau de acertos? Será minha linha de base para comparar com os modelos que serão criados.

#### Estratégias utilizadas pelo DummyClassifier:
* most_frequent  - sempre retorna o rótulo de classe mais frequente
* prior *Default - sempre retorna o rótulo de classe mais frequente
* stratified     - proporcional aos dados de entrada
* uniform        - gera previsões uniformemente aleatoriamente
* constant       - sempre prevê um rótulo constante fornecido pelo usuário

```python
def classificador_dummy(valores):
    np.random.seed(SEED)

    # Divisão dos dados entre treino e teste
    X_train, X_test, y_train, y_test = train_test_split(valores, y, test_size=0.3, stratify=y)
    
    dummy = DummyClassifier(strategy='most_frequent')
    dummy.fit(X_train, y_train)
    
    print(f'Percentual de acurácia do DummyClassifier most_frequent - {(dummy.score(X_test, y_test) * 100):.2f}%')

classificador_dummy(X)    
```

<font color='blue'>Assim, com um classificador inocente, que sempre "chuta" o valor mais frequente, alcançou uma acurácia de 62,57% que servirá de linha de base para comparação dos modelos gerados.</font>

### Avaliando com um modelo RandomForestClassifier
Modelo gera uma série de classificadores de árvores de decisão em várias subamostras do conjunto de dados e usa a média para melhorar a precisão preditiva.

Criando uma função que recebe os dados e realiza previsões, apresentando a acurácia

```python
def classificador(valores):
    # Semente para permitir repetir os teste
    np.random.seed(SEED)

    # Divisão dos dados entre treino e teste
    X_train, X_test, y_train, y_test = train_test_split(valores, y, test_size=0.3, stratify=y)

    # Random forest com 10 arvores
    clr = RandomForestClassifier(n_estimators=100)
    clr = clr.fit(X_train, y_train)
    y_pred = clr.predict(X_test)

    print("Classificador Random Forest:\n RandomForestClassifier(n_estimators=100)\n\n")

    print("--------------------------------------")
    print("Acurácia da base de treinamento: {:.2f}".format(clr.score(X_train, y_train)))
    print("--------------------------------------\n\n")
    
    print(classification_report(y_test, y_pred, target_names=y_test.unique()))
    
    cnf_matrix = confusion_matrix(y_test, y_pred)
    cnf_table = pd.DataFrame(data=cnf_matrix, index=y_test.unique(), columns=[x + "(prev)" for x in y_test.unique()])
    display(cnf_table)
    
    return clr, X_train, X_test, y_train, y_test

modelo_forest, X_train, X_test, y_train, y_test = classificador(X)
```

<font color='blue'>A acurácia de 96% legitima o Classificador Random Forest como um bom classificador, bem acima da acurácia atingida pelo DummyClassifier. Sendo assim, usaremos como base para nosso processo de redução de dimensionalidade.</font>

Criando uma função para plotar o gráfico de todos os exames

```python
def grafico_violino(ini=0, fim=10):    
    # Concatenando os dados para apresentar a visualização   
    dados_plot = pd.concat([y, X.iloc[:,ini:fim]], axis=1) 
    
    # pd.melt() vai readequar o dataframe mantendo a coluna diagnóstico e listando outras colunas com seus valores    
    dados_plot = pd.melt(dados_plot, 
                     id_vars='diagnostico',   # Colunas usada como variáveis identificadora
                     var_name='exames',       # Nome para usar na coluna variável
                     value_name='valores',    # Nome para usar na coluna valor.
                     ignore_index=False       # True-Índice original é ignorado; False-Índice original é mantido
                    )
    
    # Definir o tamanho da área plotada
    plt.figure(figsize=(15, 5))
    
    # Gera o violinplot
    sns.violinplot(
        data=dados_plot, 
        x='exames', 
        y='valores', 
        hue='diagnostico',
        split=True
    )
    
    # Rotaciona os textos do eixo X
    plt.xticks(rotation = 90)
    
    # Apresenta o gráfico
    plt.show() 
```

```python
_ = [grafico_violino(i, i+10) if (i + 10) < X.shape[1] else grafico_violino(i, X.shape[1]) for i in range(0, X.shape[1], 10)]
```

### Matriz de Correlação
O próximo objetivo é calcularmos a correlação a partir do nosso dataframe X, o que é possível com a função corr() do Pandas, que nos retornará justamente a matriz de correlação (a correlação entre todas as variáveis).

```python
X_corr = X.corr()

X_corr.iloc[:9, :9]
```

Listado a correlação para os 10 primeiros exames, porém, essa não é uma maneira agradável de fazermos uma análise desses dados. Seria mais interessante se tivéssemos uma visualização, como um mapa de calor. A biblioteca Seaborn nos disponibiliza a função heapmap() que nos possibiliza criar esse tipo de visualização.

```python
# Definir o tamanho da área plotada
plt.figure(figsize=(17, 20))
    
sns.heatmap(
    X_corr, 
    cmap='coolwarm',        # cmap - paleta de cores
    annot=True,             # annot - descrição de valor de cada quadrante
    linecolor='white',      # linecolor - cor da linha que separa os quadrante
    linewidths=2,           # linewidths - largura da linha que separa os quadrantes
    fmt='.2f'               # formato do número em cada célula, usado 2 casa decimal
)

plt.show()
```

É possível encontrar diversas features totalmente correlacionadas, com valores iguais a 1.0, e outras com um valor bem próximo, como 0.9. Enquanto isso, também temos features pouquíssimo correlacionadas, com valores próximos a 0.0. Por enquanto trabalharemos na remoção das features altamente correlacionadas.

```python
# Identificando correlações acima de 0,99 (altamente correlacionadas)
var_correlacionadas = X_corr[X_corr > 0.99].sum()
var_correlacionadas[var_correlacionadas > 1]
```

#### 3- pontuação para cada feature (cada exame) e selecionar as melhores dentre elas.
</br>
<font color=blue>SelectKBest do módulo sklearn.feature_selection é uma função matemática que gera um score para cada feature e seleciona um determinado número de features dentre os melhores scores..</font>

```python
# chi2 - função Qui-quadrado para inferir quais features serão mais representativas para nosso processo de classificação
# k = 5 exames que serão utilizados para chegar a um diagnóstico
k = 5
selecao_kmelhores = SelectKBest(score_func=chi2, k=k)

# Obs - SelectKBest não aceita valores negativos
selecao_kmelhores.fit(X_train, y_train)

# Gerando um df dos melhores para apresentação
df_kmelhores = pd.DataFrame(                     # criar DataFrame
    selecao_kmelhores.scores_,                   # scores atribuidos pelo SelectKBest
    index=selecao_kmelhores.feature_names_in_,   # nomes dos exames
    columns=['scores']                           # nome da coluna
).sort_values(                                   # método de ordenação
    by=['scores'],                               # ordenado por scores
    ascending=False                              # decrescente
)

df_kmelhores.T
```

```python
fig, ax = plt.subplots(figsize=(17, 8))
df_kmelhores.plot.bar(ax=ax)
ax.set_title("SelectKBest do módulo sklearn.feature_selection")
ax.set_ylabel("Mean accuracy decrease")
fig.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(17, 8))
sns.set_context('poster', font_scale=0.6)
sns.set_style('darkgrid')
sns.scatterplot(data= X, x=df_kmelhores[:2].index[0], y=df_kmelhores[:2].index[1], hue = y)
ax.set_title("Dois principais exames conforme modelo SelectKBest")
fig.tight_layout()
plt.show()
```

#### 4- Processo cíclico de descarte da menor acurácia.
</br>
<font color=blue>RFE (Eliminação de Feature por Recursão) do módulo sklearn.feature_selection é um método que por meio da acurácia, que será nossa forma de avaliação, nosso classificador saberá qual das features é mais importante pra ele, descartando as de menor acurácia. Esse processo será feito sucessivamente até atingir o número de features selecionado - no no nosso caso, 5.</font>

```python
# estimator - modelo
# n_features_to_select - O número de recursos para selecionar.
# step - corresponde ao número de features para remover em cada iteração. 
# verbose - controla a verbosidade da saída.
# importance_getter - string com nome/caminho do atributo para extrair a importância do recurso
selecao_RFE = RFE(estimator=modelo_forest, n_features_to_select=5, step=1)
selecao_RFE.fit(X_train, y_train)
```

```python
# Gerando um df dos melhores para apresentação
df_RFE = pd.DataFrame(                           # criar DataFrame
    selecao_RFE.ranking_,                        # scores atribuidos pelo RFE
    index=selecao_RFE.feature_names_in_,         # nomes dos exames
    columns=['scores']                           # nome da coluna
).sort_values(                                   # método de ordenação
    by=['scores'],                               # ordenado por scores
    ascending=True                               # Os recursos de melhor estimado são atribuídos ao rank 1
)

df_RFE.T
```

```python
print(f'O ranking de recursos que corresponde à posição de classificação do recursos, de melhor estimado, são atribuídos ao rank 1.')

fig, ax = plt.subplots(figsize=(17, 8))
df_RFE.plot.bar(ax=ax)
ax.set_title("RFE do módulo sklearn.feature_selection")
ax.set_ylabel("Ranking")
fig.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(17, 8))
sns.set_context('poster', font_scale=0.6)
sns.set_style('darkgrid')
sns.scatterplot(data= X, x=df_RFE[:2].index[0], y=df_RFE[:2].index[1], hue = y)
ax.set_title("Dois principais exames conforme modelo SelectKBest")
fig.tight_layout()
plt.show()
```

#### 5- Técnica de eliminação de recursos recursivos com validação cruzada para selecionar o número de recursos
</br>
<font color=blue>RFECV (RFE Cross Validation) - Divide o nosso banco de dados em blocos e aplica o algoritmo RFE, que acabamos de aprender, em cada um desses blocos, gerando diferentes resultados. Dessa forma, O RFECV não só nos informa quantas features precisamos ter para gerar o melhor resultado possível, como também quais features são essas.</font>

```python
# estimator - modelo
# cv - Determina a estratégia de divisão de validação cruzada.
# step - corresponde ao número de features para remover em cada iteração. 
# verbose - controla a verbosidade da saída.
# importance_getter - string com nome/caminho do atributo para extrair a importância do recurso
selecao_RFECV = RFECV(estimator=modelo_forest, cv=5, step=1, scoring='accuracy')
selecao_RFECV.fit(X_train, y_train)
```

```python
# Gerando um df dos melhores para apresentação
df_RFECV = pd.DataFrame(                           # criar DataFrame
    selecao_RFECV.ranking_,                      # scores atribuidos pelo RFE
    index=selecao_RFECV.feature_names_in_,       # nomes dos exames
    columns=['scores']                           # nome da coluna
).sort_values(                                   # método de ordenação
    by=['scores'],                               # ordenado por scores
    ascending=True                               # Os recursos de melhor estimado são atribuídos ao rank 1
)

exames_selecionados = df_RFECV[df_RFECV.scores == 1].index

print(f'RFECV (RFE Cross Validation) recomendou {len(exames_selecionados)} exames')

', '.join(exames_selecionados)
```

```python
print(f'O ranking de recursos que corresponde à posição de classificação do recursos, de melhor estimado, são atribuídos ao rank 1.')

fig, ax = plt.subplots(figsize=(17, 8))
df_RFECV.plot.bar(ax=ax)
ax.set_title("RFECV do módulo sklearn.feature_selection")
ax.set_ylabel("Ranking")
fig.tight_layout()
plt.show()
```

#### 6- Técnica de para manter o máximo de informações possível em uma dimensionalidades reduzidas.
</br>
<font color=blue>
    PCA (Análise de componentes principais) - opera uma transformação matemática sobre os dados, na qual as primeiras dimensões serão as variáveis que manterão o maior volume possível de informação.</br></br>
    Essas técnicas que realizam uma transformação matemática nos dados antes de processá-los, reduzindo suas dimensões, são chamadas justamente de "redução de dimensionalidade". A diferença entre a redução de dimensionalidade e a seleção de features é que a segunda funciona como um funil que filtra as principais características de uma base de dados original.
</font>

#### 7- Cálculo entre as distancias nas 28 dimensões.
</br>
<font color=blue>
    t-SNE é uma ferramenta para visualizar dados de alta dimensão. calcula a distância entre os pontos, teremos <strong>um valor em 28 dimensões</strong>, que é a quantidade de features do nosso conjunto. O t-SNE tenta manter a proporção dessas distâncias na redução para dimensões menores, permitindo que tenhamos uma noção de quão distante um ponto está do outro.</br></br>
    É altamente recomendável usar outro método de redução de dimensionalidade (por exemplo, PCA) para reduzir o número de dimensões a uma quantidade razoável se o número de registros for muito alto. Isso suprimirá algum ruído e acelerará o cálculo de distâncias entre as amostras.
</font>

```python
# Transformação TSNE()
tsne = TSNE(n_components = 2, random_state=SEED)
X_tsne = tsne.fit_transform(X)

# Aprentação da visualização
fig, ax = plt.subplots(figsize=(17, 8))
sns.set_context('poster', font_scale=0.6)
sns.set_style('darkgrid')
sns.scatterplot(x=X_tsne[:,0], y=X_tsne[:,1], hue = y)
ax.set_title("Visualização das distancias entre os pontos em 28 dimensões")
plt.show()
```

### <font color='red'>Importante</font>
<font color=red>
    Ao se utilizar as técnicas de seleção de feature para gerar uma visualização no plano, é preciso escolher dois dos atributos para representar os dados. Manter apenas dois dos diversos atributos, faz com que um grande número de informação seja perdida, resultado em visualizações que podem induzir à conclusões equivocadas.</br>
    Para gerar uma visualização no plano, o ideal é utilizar técnicas que tentam preservar o maior número possível de informação em uma dimensão reduzida, como PCA e T-SNE.
