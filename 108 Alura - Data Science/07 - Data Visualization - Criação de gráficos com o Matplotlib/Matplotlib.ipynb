### dtypes: datetime
Percebemos que o conjunto consiste em mais de 45 mil linhas, nenhuma delas nula. Além disso, o campo data é do tipo object, que é interpretado pelo Python como uma string. Entretanto, é mais interessante para nós que a variável de data seja interpretada como o tipo datetime, o que nos ajudará na construção dos gráficos. Consertaremos esse ponto atribuindo à coluna df['data'] a conversão dessa mesma coluna no tipo desejado, algo que faremos usando a <b>função to_datetime()</b> do Pandas.

```python
df_tempo['data'] = pd.to_datetime(df_tempo['data'])
```

### Linhas de restrição e anotações

```python
max(df_tempo['temperatura'])
```

```python
fig = plt.figure(figsize=(13, 6))

eixo = fig.add_axes([0, 0, 1, 1])

eixo.plot(df_tempo['data'], df_tempo['temperatura'], color='cadetblue')

eixo.set_title('Temperatura na escala Kelvin no Tempo', pad=20, fontsize=26)
eixo.set_xlabel('Data', fontsize=15)
eixo.set_ylabel('Temperatura', fontsize=15)

eixo.legend(['Temperatura'], loc='lower right', fontsize=13)

# Linhas de restrição
eixo.axhline(max(df_tempo['temperatura']), color='k', linestyle='--')
eixo.axhline(min(df_tempo['temperatura']), color='k', linestyle='--')

# anotações
# idxmax() para pegar o indice do maior valor de temperatura
x1 = df_tempo['data'][df_tempo['temperatura'].idxmax()]
y1 = max(df_tempo['temperatura'])
eixo.annotate("Máximo", xy=(x1, y1), fontsize=20)

eixo.grid(True)
```

```python
fig = plt.figure(figsize=(13, 6))

eixo = fig.add_axes([0, 0, 1, 1])

eixo.plot(df_tempo['data'], df_tempo['temperatura'], color='cadetblue')

eixo.set_title('Temperatura na escala Kelvin no Tempo', pad=20, fontsize=26)
eixo.set_xlabel('Data', fontsize=15)
eixo.set_ylabel('Temperatura', fontsize=15)

eixo.legend(['Temperatura'], loc='lower right', fontsize=13)

# Linhas de restrição
eixo.axhline(max(df_tempo['temperatura']), color='k', linestyle='--')
eixo.axhline(min(df_tempo['temperatura']), color='k', linestyle='--')

# anotações (max) ----------------------------------------------------------------------------------------------------------------------
# idxmax() para pegar o indice do maior valor de temperatura
destino_x = df_tempo['data'][df_tempo['temperatura'].idxmax()]
destino_y = max(df_tempo['temperatura'])

origem_x  = df_tempo['data'][df_tempo['temperatura'].idxmax() - 7000] # - 7000 horas
origem_y  = max(df_tempo['temperatura'] - 15)

eixo.annotate("Máximo", xy=(destino_x, destino_y), fontsize=20, xytext=(origem_x, origem_y), arrowprops=dict(facecolor='k'))

# anotações (min) ----------------------------------------------------------------------------------------------------------------------
# idxmax() para pegar o indice do maior valor de temperatura
destino_x = df_tempo['data'][df_tempo['temperatura'].idxmin()]
destino_y = min(df_tempo['temperatura'])

origem_x  = df_tempo['data'][df_tempo['temperatura'].idxmin() + 3500] # - 3500 horas
origem_y  = min(df_tempo['temperatura'] + 10)

eixo.annotate("Mínimo", xy=(destino_x, destino_y), fontsize=20, xytext=(origem_x, origem_y), arrowprops=dict(facecolor='k'))

eixo.grid(True)
```

```python
fig = plt.figure(figsize=(5, 4))

eixo = fig.add_axes([0, 0, 1, 1])

indice = range(len(temperatura_media_por_dia_da_semana))

# Explodir para destacar a fatia da pizza, passando o valor entre 0 e 1
explodir = [0.1, 0, 0, 0, 0, 0, 0.1]

# autopct='%.1f%%' aplica o percentual para cada fatia
# shadow aplica um sombra na fatia
eixo.pie(temperatura_media_por_dia_da_semana, labels=temperatura_media_por_dia_da_semana.index,
        autopct='%.1f%%', explode=explodir, shadow=True)

eixo.set_title('Temperatura Média Por Dia da Semana', fontsize=18, pad=10)
```

```python
fig = plt.figure(figsize=(15, 8))

eixo = fig.add_axes([0, 0, 1, 1])

# Atribuindo cores para cada especie
cores = {'Iris-setosa': 'r', 'Iris-versicolor': 'b', 'Iris-virginica': 'g'}

# Diferenciando marcadores
marcadores = {'Iris-setosa': 'x', 'Iris-versicolor': 'o', 'Iris-virginica': 'v'}

# Necessário plotar para cada cor
# s=100 para tamanho do marcador
for especie in df_iris['espécie'].unique():
    tmp = df_iris[df_iris['espécie'] == especie]
    eixo.scatter(tmp['comprimento_sépala'], tmp['largura_sépala'], 
                 color=cores[especie], marker=marcadores[especie], s=100)
    
eixo.set_title('Gráfico de Dispersão', fontsize=25, pad=15)
eixo.set_xlabel('Comprimento da Sépala', fontsize=15)
eixo.set_ylabel('Largura da Sépala', fontsize=15)

# labelsize recebe o tamanho dos número nos eixos X e Y.
eixo.tick_params(labelsize=20)

eixo.legend(cores, fontsize=20)

eixo.grid(True)
```

# Aula 04.4
### Visualizando a distribuição de dados com Box-Plot

```python
fig = plt.figure(figsize=(8, 4))

eixo = fig.add_axes([0, 0, 1, 1])

cores = ['red', 'blue', 'orange', 'green']

# patch_artist=True para preencher com cor os box-plot
caixas = eixo.boxplot(df_iris.drop('espécie', axis='columns'), patch_artist=True)

eixo.set_title('Gráfico de Caixa - Box-Plot', fontsize=15, pad=10)
eixo.set_xticklabels(df_iris.drop('espécie', axis='columns').columns)

# Marcadores de outliers como 'x'
for outlier in caixas['fliers']:
    outlier.set(marker='x', markersize=8)

# iterando para cada caixa e definindo as cores
for caixa, cor in zip(caixas['boxes'], cores):
    caixa.set(color=cor)
```

```python
fig = plt.figure(figsize=(10, 4))

eixo = fig.add_axes([0, 0, 1, 1])

eixo.hist(df_iris['comprimento_pétala'], bins=20, orientation='vertical')

eixo.set_title('Histograma', fontsize=15, pad=10)
eixo.set_xlabel('Comprimento da Pétala', fontsize=15)

# mu - média
# sigma - desvio padrão
mu, sigma = df_iris['comprimento_pétala'].mean(), df_iris['comprimento_pétala'].std()

eixo.annotate(f'$\mu = {mu:.2f}$\n$\sigma = {sigma:.2f}$', xy=(6.1, 25), fontsize=20)

# Linha vertical com a média
eixo.axvline(mu, color='k', linestyle='--')
eixo.annotate('média', xy=(3.1, 26), fontsize=20)

# Linha vertical com a mediana
eixo.axvline(df_iris['comprimento_pétala'].median(), color='g', linestyle='--')
eixo.annotate('mediana', xy=(4.4, 22), fontsize=20, color='g')

eixo.grid(True)
```

Importando a biblioteca PIL, utilizada para manipular imagens.

```python
from PIL import Image
```

```python
imagens = ['histograma_versicolor.png', 'histograma_virginica.png', 'histograma_setosa.png']

largura_max, altura_tot, intervalo = 0, 0, 0

# Calculando largura máxima é altura total
for imagem in map(Image.open, imagens):
    largura, altura = imagem.size
    
    print(f'A imagem {imagem.filename} possui largura de {largura}px e altura de {altura}px.')
    
    largura_max = largura if largura > largura_max else largura_max
    altura_tot += altura

print(f'\nLargura máxima {largura_max}px e altura total {altura_tot}px.\n')

# Criando uma imagem combinada
combinado = Image.new('RGB', (largura_max, altura_tot))

# Combinando as imagens salvas numa só figura
for imagem in map(Image.open, imagens):
    combinado.paste(imagem, (0, intervalo))
    
    largura, altura = imagem.size
    intervalo += altura

# Salvando a imagem combinada em disco
combinado.save('combinado.png')
```
